```python
import os
import zipfile

def unzip_subject_zips(folder_path, save_path):
    subject_folder = os.path.abspath(folder_path)
    subject_zips = [file for file in os.listdir(subject_folder) if file.endswith('.zip')]

    for zip_file in subject_zips:
        zip_file_path = os.path.join(subject_folder, zip_file)
        folder_name = os.path.splitext(zip_file)[0]
        folder_path = os.path.join(save_path, folder_name)

        if not os.path.exists(folder_path):
            os.makedirs(folder_path)
            with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
                zip_ref.extractall(folder_path)
        else:
            print(f"Folder '{folder_name}' already exists. Skipping extraction.")

if __name__ == "__main__":
    subject_folder_path = "/mnt/HDD3/btan8779/Data/Explorer"
    save_path = "/mnt/HDD3/btan8779/Challenge_Data/Explorer2"
    unzip_subject_zips(subject_folder_path, save_path)
```

```python
def SUV_with_decay(pet_file,PET_data):
    # pet_file = pydicom.dcmread(Image_File_PATH)
    # print(pet_file)
    # PET_data = sitk.ReadImage(PET_File_PATH)
    PET_array = sitk.GetArrayFromImage(PET_data).astype(np.float32)
    # pet_file_data = pet_file.pixel_array.astype(np.float32)
    slope = pet_file[0x0028, 0x1053].value
    intercept = pet_file[0x0028, 0x1052].value
    # activity_concentration = PET_array * slope + intercept
    activity_concentration = PET_array
    
    # tmp = pet_file[0x0054,0x0016].value
    # print(tmp)
    injected_dose = pet_file[0x0054,0x0016][0][0x0018,0x1074].value
    half_time = pet_file[0x0054,0x0016][0][0x0018, 0x1075].value
    
    # don't consider this situation: the series date and the start data are not the same day
    series_date = pet_file[0x0008, 0x0031].value
    start_date = pet_file[0x0054,0x0016][0][0x0018, 0x1072].value
    # if series_date != start_date:
    #     print(': the PET series date is not the same as the start_date, Pay Attention!!!')
    time = pet_file[0x0008, 0x0031].value
    time_hour = float(time[0:2])
    time_minute = float(time[2:4])
    time_second = float(time[4:6])
    
    # start_time = pet_file[0x0054,0x0016][0][0x0018, 0x1072].value
    start_time = start_date
    start_hour = float(start_time[0:2])
    start_minute = float(start_time[2:4])
    start_second = float(start_time[4:6])
    
    delta_t = (time_hour - start_hour) * 3600 + (time_minute - start_minute) * 60 + (time_second - start_second)
    decay_correction = pow(2, (- delta_t / half_time))
    
    weight = pet_file[0x0010, 0x1030].value * 1000
    
    pet_file_data = activity_concentration * weight / (injected_dose * decay_correction)
    print('patient weight : ', weight )
    print('patient injected dose :', injected_dose)

    slice_num = pet_file_data.shape[0]
    max_suv = []
    # for index in labeled_indexes:
    #     c = pet_file_data[index, :, ]
    #     labeled_pixel_0 = labeled_pixel[index]
    #     labeled_c = c[labeled_pixel_0]
    #     # print(max(labeled_c))
    #     max_suv.append(np.mean(labeled_c))
    for slice in range(slice_num):
        # print(slice)
        suv_slice =pet_file_data[slice, :, ]
        max_suv.append(np.max(suv_slice))
    
    max_suv= np.array(max_suv)
    print('patient suv ma',np.mean(max_suv))
    suv_image = sitk.GetImageFromArray(pet_file_data)
    return suv_image
```

```python
def nii_file_save(save_path, nii_data):
    wriiter = sitk.ImageFileWriter()
    wriiter.SetFileName(save_path)
    wriiter.Execute(nii_data)
```

```python
import os

def rename_folders_sync(path_a, path_b):
    counter = 1
    for folder_a in os.listdir(path_a):
        a_path = os.path.join(path_a,folder_a)
        folder_list = os.listdir(a_path)
        for folder in folder_list:
            if 'Subject' in folder:
                folder_path = os.path.join(path_a,'Subject',folder)
            elif 'Explorer' in folder:
                folder_path = os.path.join(path_a,'Explorer',folder)
            else:
                print('wrong')
                pass
            patient_list = os.listdir(folder_path)
            
            # print(patient_list)
            for patient in patient_list:
                patient_path_a = os.path.join(folder_path,patient)
                patient_path_b = os.path.join(path_b,patient)
                if os.path.isdir(patient_path_a) and os.path.exists(patient_path_b):
                    new_name_in_a = os.path.join(a_path, f"patient{counter}")
                    new_name_in_b = os.path.join(path_b, f"patient{counter}")

                    os.rename(patient_path_a, new_name_in_a)
                    os.rename(patient_path_b, new_name_in_b)

                    print(f"Renamed: {patient} -> patient{counter}")
                    counter += 1
                else:
                    print('wrong')
                    
```

```python
import os
import re

def natural_sort_key(s):
    for text in re.split(r'(\d+)',s):
        if text.isdigit():
            return int(text)
        else:
            text.lower()

def rename_folders(folder_path, prefix, padding=3):
    folder_names = sorted(os.listdir(folder_path), key=natural_sort_key)
    for index, folder_name in enumerate(folder_names,1):
        if os.path.isdir(os.path.join(folder_path, folder_name)):
            
            new_folder_name = f"{prefix}{str(index).zfill(padding)}"
            # print(new_folder_name,folder_name)
            os.rename(
                os.path.join(folder_path, folder_name),
                os.path.join(folder_path, new_folder_name)
            )
            print(f"Renamed {folder_name} to {new_folder_name}")

folder_path = data_path_explorer  # Replace with the actual path to the parent folder
prefix = "patient"
padding = 3  # Adjust padding as needed (e.g., 2 for patient01, patient02, etc.)

rename_folders(folder_path, prefix, padding)
```
